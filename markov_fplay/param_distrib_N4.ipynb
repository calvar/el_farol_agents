```python
nr = len(dfucb)
nc = 2
fig0, ax = plt.subplots(nrows=nr,ncols=nc,figsize=(5,10))
for i in range(len(dfucb)):
    pa = sns.histplot(data=dfucb[i], ax=ax[i,0], x='efficiency', binwidth=10)
    pb = sns.histplot(data=dfucb[i], ax=ax[i,1], x='inequality', binwidth=10)
    pa.set_xlim(0,500)
    pa.set_ylim(0,200)
    pb.set_xlim(0,500)
    pb.set_ylim(0,200)
plt.tight_layout()
```

```python
cvals = [0,0.01,0.05,0.08,0.1,0.3,0.5,0.7,1.0] 
eff_means = [d.efficiency.mean() for d in dfucb]
inq_means = [d.inequality.mean() for d in dfucb]
plt.plot(cvals,eff_means)
plt.plot(cvals,inq_means)
```

```python
import numpy as np
from numpy.fft import fft, fftshift, fftfreq
import matplotlib.pyplot as plt
```

```python
def bin_to_dec(b_str):
    if len(b_str) > 1:
        return int(b_str[0])*2**(len(b_str)-1) + bin_to_dec(b_str[1:])
    else:
        return int(b_str[0])
```

```python
#Read state sequence
seq = []
with open("data/state_evol_0030.dat", "r") as fdata:
    for line in fdata:
        v = line.split()
        seq.append(bin_to_dec(v[1]))
seq = np.array(seq)
print(seq)
```

```python
fig1, ax = plt.subplots(figsize=(6,6))
ax.plot(range(len(seq)),seq)
ax.set_xlim([150,200])
```

```python
freq = fftfreq(len(seq))*len(seq)
ft = fft(seq)
```

```python
fig1, ax = plt.subplots(figsize=(6,6))
ax.plot(freq,np.imag(ft))
```

```python
with open("serie1.csv", "w") as f:
    for i in seq:
        f.write("{0}, ".format(i))
```

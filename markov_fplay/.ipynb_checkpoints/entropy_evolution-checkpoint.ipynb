```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder = 'data/data_N16C0.01'
maxN = 3001
N = 16
```

```python
def bin_to_dec(b_str):
    if len(b_str) > 1:
        return int(b_str[0])*2**(len(b_str)-1) + bin_to_dec(b_str[1:])
    else:
        return int(b_str[0])
```

```python
#Read state sequences
state_bins = []
state_data = []
for n in range(0,maxN,5):
    states = []
    with open(folder+"/state_evol_{0:04d}.dat".format(n), "r") as fdata:
        for line in fdata:
            v = line.split()
            states.append(bin_to_dec(v[1]))
    state_data.append(states)
state_data = np.array(state_data)
state_data.shape
```

```python
def entropy(freq):
    tot = 0
    for i in freq:
        tot += freq[i]
    dist = {}
    for i in freq:
        dist[i] = freq[i]/tot
    H = 0
    for i in dist:
        if dist[i] > 0:
            H -= dist[i] * np.log2(dist[i])
    return H
```

```python
entropy_evols = []
for n in range(state_data.shape[0]):
    entropy_ev = []
    frequency = {}
    for i in range(state_data.shape[1]):
        frequency[state_data[n,i]] = frequency.get(state_data[n,i],0) + 1
        entropy_ev.append( entropy(frequency)/N )
    entropy_evols.append(entropy_ev)
```

```python
fig = plt.figure(figsize = (6,6))
for i in range(len(entropy_evols)):
    plt.plot(entropy_evols[i])
plt.xlabel('step')
plt.ylabel('entropy')
```
